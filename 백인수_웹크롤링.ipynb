```python
import requests
```

```python
import bs4
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.daum.net/premovie/netflix'
```

```python
res = requests.get(url)
```

```python
with open('백인수_넷플릭스.html','w',encoding='utf-8') as fp:
    fp.write(res.text)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
ul = soup.find('ul', class_='list_movieranking aniposter_ott')
```

```python
img_src = ul.find_all('a')
```

```python
print(len(img_src))
```

```python
for c in img_src:
    img_src1 = c['href']
    print(img_src1)
```

```python
with open('백인수_result.txt','w',encoding='utf-8') as fp: 
    print(f'검색된 이미지는 총 {len(img_src)}건 입니다.')
    fp.write(f'검색된 이미지는 총 {len(img_src)}건 입니다.\n')
    for i in range(len(img_src)):
        print(f'https://movie.daum.net{img_src1}')
        fp.write(f'https://movie.daum.net{img_src1}')
```
